```python
import seaborn as sns
import matplotlib.pyplot as plt
import math

columns_to_look = ["Sex", "Marital", "Race", "Albuminuria", "MetabolicSyndrome"]

# Grid setup: 2 plots per row
n_cols = 2
n_rows = math.ceil(len(columns_to_look) / n_cols)

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(9, 4 * n_rows))
axes = axes.flatten()
pie_train_df = train_df
for i, col in enumerate(columns_to_look):
    non_null_values = pie_train_df[col].dropna()
    value_counts = non_null_values.value_counts()
    labels = value_counts.index.tolist()
    pie_colors = sns.color_palette('Set2', len(labels))

    axes[i].pie(
        value_counts,
        labels=labels,
        shadow=True,
        explode=[0.05] * len(labels),
        autopct='%1.1f%%',
        textprops={'size': 12},
        colors=pie_colors,
        startangle=90
    )
    axes[i].set_title(f'{col} Distribution', fontsize=14)

# Hide unused axes if the number of columns is odd
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def plot_numeric_kde_boxplot(train_df, numeric_cols):
    # Define a list of Seaborn color palettes
    palette_list = ['Set1', 'Set2', 'Set3', 'tab10', 'Paired', 'Dark2', 'pastel']

    for i, col in enumerate(numeric_cols):
        fig, axes = plt.subplots(1, 2, figsize=(16, 5))
        
        # Pick a palette based on index (cyclic)
        palette = sns.color_palette(palette_list[i % len(palette_list)])
        color_hist = palette[0]
        color_box = palette[1] if len(palette) > 1 else palette[0]

        # KDE + Histogram
        sns.histplot(train_df, x=col, kde=True, color=color_hist, ax=axes[0])
        axes[0].set_title(f'{col} - Histogram & KDE', fontsize=14)

        # Boxplot
        sns.boxplot(train_df, y=col, color=color_box, ax=axes[1])
        axes[1].set_title(f'{col} - Boxplot', fontsize=14)

        # Overall title
        fig.suptitle(f'Numeric Feature: {col}', fontsize=16, fontweight='bold')
        fig.subplots_adjust(wspace=0.3)
        plt.tight_layout()
        plt.show()


numeric_cols = train_df.drop(columns=["seqn", "Albuminuria","MetabolicSyndrome"]).select_dtypes(include="number").columns
plot_numeric_kde_boxplot(train_df, numeric_cols)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import QuantileTransformer
transform_train_df = train_df
transform_test_df = test_df
# Variables to transform
variables = ["WaistCirc", "BMI", "UrAlbCr", "UricAcid", "BloodGlucose", "HDL", "Triglycerides"]

# Initialize transformers (fit only on train_df later)
quantile_tf = QuantileTransformer(output_distribution='normal', random_state=42)

# Store transformed test data (optional)
test_transformed = pd.DataFrame()

for var in variables:
    # Create figure
    fig, axes = plt.subplots(1, 3, figsize=(18, 4))
    
    # --- Training Data ---
    train_data = transform_train_df[var].dropna()
    
    # Apply transformations to TRAIN data
    log_trans_train = np.log1p(train_data)
    sqrt_trans_train = np.sqrt(train_data)
    quantile_trans_train = pd.Series(
        quantile_tf.fit_transform(train_data.values.reshape(-1, 1)).flatten(),  # Fit here
        index=train_data.index
    )
    
    # --- Test Data (transformed using train-fit transformers) ---
    test_data = transform_test_df[var].dropna()
    quantile_trans_test = pd.Series(
        quantile_tf.transform(test_data.values.reshape(-1, 1)).flatten(),  # No fit!
        index=test_data.index
    )
    # Store test transforms if needed
    test_transformed[f'{var}_quantile'] = quantile_trans_test
    
    # --- Plot TRAIN distributions ---
    sns.histplot(log_trans_train, ax=axes[0], kde=True, color='darkorange')
    axes[0].set_title(f'Train: {var} (Log)')
    
    sns.histplot(sqrt_trans_train, ax=axes[1], kde=True, color='seagreen')
    axes[1].set_title(f'Train: {var} (Sqrt)')
    
    sns.histplot(quantile_trans_train, ax=axes[2], kde=True, color='slateblue')
    axes[2].set_title(f'Train: {var} (Quantile)')
    
    fig.suptitle(f'Training Set Transformations: {var}', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()

# Optional: Add transformed test data to test_df
for var in variables:
    transform_test_df[f'{var}_quantile'] = test_transformed[f'{var}_quantile']
```

```python
import pandas as pd
from sklearn.preprocessing import QuantileTransformer
transform_train_df = train_df 
# Variables to transform
variables = ["WaistCirc", "BMI", "UrAlbCr", "UricAcid", "BloodGlucose", "HDL", "Triglycerides"]

# Initialize transformer (fit only on train data)
quantile_tf = QuantileTransformer(output_distribution='normal', random_state=42)

# --- Transform TRAIN data ---
# Fit and transform training data
train_quantile = pd.DataFrame(
    quantile_tf.fit_transform(transform_train_df[variables]),
    columns=[f"{var}_quantile" for var in variables],
    index=transform_train_df.index
)

# --- Transform TEST data ---
# Use same transformer (no fitting)
transform_test_df = test_df
test_quantile = pd.DataFrame(
    quantile_tf.transform(transform_test_df[variables]),
    columns=[f"{var}_quantile" for var in variables],
    index=transform_test_df.index
)

# --- Update DataFrames ---
# 1. Drop original variables
transform_train_df = transform_train_df.drop(columns=variables)
transform_test_df = transform_test_df.drop(columns=variables)
train_df = train_df.drop(columns=variables)
test_df = test_df.drop(columns=variables)

# 2. Add transformed versions
transform_train_df = pd.concat([transform_train_df, train_quantile], axis=1)
transform_test_df = pd.concat([transform_test_df, test_quantile], axis=1)

# Verify
print("Train columns:", transform_train_df.columns.tolist())
print("Test columns:", transform_test_df.columns.tolist())
print("Train columns:", train_df.columns.tolist())
print("Test columns:", test_df.columns.tolist())
```
