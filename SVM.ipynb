```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.metrics import (accuracy_score, recall_score, f1_score, 
                            roc_auc_score, confusion_matrix, classification_report)
import matplotlib.pyplot as plt
import seaborn as sns

# =============================================
# 1. DATA PREPARATION
# =============================================

# Separate features and target (assuming seqn is index)
X_train = train_set_clean_transform.drop(['MetabolicSyndrome','seqn','QT_UrAlbCr','QT_WaistCirc'], axis=1)
y_train = train_set_clean_transform['MetabolicSyndrome'].cat.codes  # Convert to 0/1

X_test = test_set_clean_transform.drop(['MetabolicSyndrome','seqn','QT_UrAlbCr','QT_WaistCirc'], axis=1)
y_test = test_set_clean_transform['MetabolicSyndrome'].cat.codes

# =============================================
# 2. PREPROCESSING SETUP
# =============================================

# Identify column types
categorical_cols = X_train.select_dtypes(include='category').columns.tolist()
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Separate QT and non-QT numerical variables
non_qt_numeric = [col for col in numeric_cols if not col.startswith('QT_')]
qt_numeric = [col for col in numeric_cols if col.startswith('QT_')]

# Create transformers
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())  # Standardize non-QT variables
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False, drop='first'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, non_qt_numeric),
        ('qt_num', 'passthrough', qt_numeric),  # Leave QT_* variables as-is
        ('cat', categorical_transformer, categorical_cols)
    ])

# =============================================
# 3. SVM MODEL PIPELINE
# =============================================

svm_model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', SVC(
        kernel='rbf',          # Radial Basis Function kernel
        C=1.0,                # Regularization parameter
        gamma='scale',        # Kernel coefficient
        probability=True,     # Enable probability estimates for AUC
        random_state=42,
        class_weight='balanced'  # Handle class imbalance
    ))
])

# Train model
svm_model.fit(X_train, y_train)

# =============================================
# 4. EVALUATION METRICS
# =============================================

def evaluate_model(model, X, y, dataset_name):
    y_pred = model.predict(X)
    y_proba = model.predict_proba(X)[:, 1] if hasattr(model, 'predict_proba') else [0]*len(y)
    
    # Calculate metrics
    metrics = {
        'Accuracy': accuracy_score(y, y_pred),
        'Sensitivity (Recall)': recall_score(y, y_pred),
        'Specificity': recall_score(y, y_pred, pos_label=0),
        'F1 Score': f1_score(y, y_pred),
        'AUC-ROC': roc_auc_score(y, y_proba) if hasattr(model, 'predict_proba') else 'N/A'
    }
    
    # Confusion matrix
    cm = confusion_matrix(y, y_pred)
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['No', 'Yes'], 
                yticklabels=['No', 'Yes'])
    plt.title(f'Confusion Matrix ({dataset_name})')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    
    print(f"\nClassification Report ({dataset_name}):")
    print(classification_report(y, y_pred, target_names=['No MetabolicSyndrome', 'MetabolicSyndrome']))
    
    return metrics

# Evaluate on both sets
train_metrics = evaluate_model(svm_model, X_train, y_train, 'Training')
test_metrics = evaluate_model(svm_model, X_test, y_test, 'Testing')

# Display results
print("\nTraining Metrics:")
print(pd.DataFrame.from_dict(train_metrics, orient='index', columns=['Value']))

print("\nTesting Metrics:")
print(pd.DataFrame.from_dict(test_metrics, orient='index', columns=['Value']))

from sklearn.inspection import permutation_importance

from sklearn.inspection import permutation_importance

# Get the fitted preprocessor from the pipeline
preprocessor = svm_model.named_steps['preprocessor']

# 1. Get feature names from StandardScaler (non-QT numeric features)
num_features = preprocessor.named_transformers_['num'].named_steps['scaler'].get_feature_names_out(non_qt_numeric)

# 2. QT_* numeric features are passed through â€” keep their names directly
qt_features = qt_numeric  # These are not transformed, just passed

# 3. Get feature names from OneHotEncoder (categorical features)
cat_features = preprocessor.named_transformers_['cat'].named_steps['onehot'].get_feature_names_out(categorical_cols)

# 4. Combine all
final_feature_names = np.concatenate([num_features, qt_features, cat_features])

# Check lengths against permutation importance
print("Total feature names:", len(final_feature_names))
print("Total importances:", len(result.importances_mean))

# No assertion â€” now using only matched length
min_len = min(len(final_feature_names), len(result.importances_mean))

# Create importance DataFrame
importance_df = pd.DataFrame({
    'Feature': final_feature_names[:min_len],
    'Importance Mean': result.importances_mean[:min_len],
    'Importance Std': result.importances_std[:min_len]
}).sort_values(by='Importance Mean', ascending=False)

# Plot top 20
plt.figure(figsize=(10, 6))
sns.barplot(data=importance_df.head(20), x='Importance Mean', y='Feature', palette='viridis')
plt.title('Top 20 Feature Importances (Permutation Importance)')
plt.xlabel('Mean Decrease in Accuracy')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()

# Print top 10 important features
print("\nTop 10 Important Features:")
print(importance_df.head(10))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import (accuracy_score, recall_score, f1_score, 
                            roc_auc_score, confusion_matrix, classification_report,
                            make_scorer)
import matplotlib.pyplot as plt
import seaborn as sns

# =============================================
# 1. DATA PREPARATION
# =============================================

# Separate features and target (assuming seqn is index)
X_train = train_set_clean_transform.drop(['MetabolicSyndrome','seqn','QT_BMI'], axis=1)
y_train = train_set_clean_transform['MetabolicSyndrome'].cat.codes  # Convert to 0/1

X_test = test_set_clean_transform.drop(['MetabolicSyndrome','seqn','QT_BMI'], axis=1)
y_test = test_set_clean_transform['MetabolicSyndrome'].cat.codes

# =============================================
# 2. PREPROCESSING SETUP
# =============================================

# Identify column types
categorical_cols = X_train.select_dtypes(include='category').columns.tolist()
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Separate QT and non-QT numerical variables
non_qt_numeric = [col for col in numeric_cols if not col.startswith('QT_')]
qt_numeric = [col for col in numeric_cols if col.startswith('QT_')]

# Create transformers
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())  # Standardize non-QT variables
])

# Using RobustScaler for QT variables (preserves quantile transformation effects)
qt_transformer = Pipeline(steps=[
    ('robust', RobustScaler())  # Better for previously quantile-transformed data
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False, drop='first'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, non_qt_numeric),
        ('qt_num', qt_transformer, qt_numeric),
        ('cat', categorical_transformer, categorical_cols)
    ])

# =============================================
# 3. SVM MODEL WITH HYPERPARAMETER TUNING
# =============================================

# Base pipeline
svm_pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', SVC(probability=True, random_state=42, class_weight='balanced'))
])

# Parameter grid
param_grid = {
    'classifier__C': [0.1, 1, 10, 100],            # Regularization strength
    'classifier__gamma': ['scale', 'auto', 0.01, 0.1, 1],  # Kernel coefficient
    'classifier__kernel': ['rbf', 'poly', 'sigmoid'],      # Kernel type
    'classifier__degree': [2, 3, 4]                        # For poly kernel
}

# Custom scoring metrics
scoring = {
    'Accuracy': make_scorer(accuracy_score),
    'Sensitivity': make_scorer(recall_score),
    'Specificity': make_scorer(recall_score, pos_label=0),
    'AUC': 'roc_auc',
    'F1': make_scorer(f1_score)
}

# Grid search with cross-validation
grid_search = GridSearchCV(
    estimator=svm_pipe,
    param_grid=param_grid,
    cv=5,
    scoring=scoring,
    refit='AUC',  # Optimize for AUC
    n_jobs=-1,
    verbose=2
)

# Run grid search
grid_search.fit(X_train, y_train)

# Best model
best_svm = grid_search.best_estimator_

# =============================================
# 4. EVALUATION
# =============================================

def evaluate_model(model, X, y, dataset_name):
    y_pred = model.predict(X)
    y_proba = model.predict_proba(X)[:, 1]
    
    # Calculate metrics
    metrics = {
        'Accuracy': accuracy_score(y, y_pred),
        'Sensitivity (Recall)': recall_score(y, y_pred),
        'Specificity': recall_score(y, y_pred, pos_label=0),
        'F1 Score': f1_score(y, y_pred),
        'AUC-ROC': roc_auc_score(y, y_proba)
    }
    
    # Confusion matrix
    cm = confusion_matrix(y, y_pred)
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['No', 'Yes'], 
                yticklabels=['No', 'Yes'])
    plt.title(f'Confusion Matrix ({dataset_name})')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    
    print(f"\nClassification Report ({dataset_name}):")
    print(classification_report(y, y_pred, target_names=['No MetabolicSyndrome', 'MetabolicSyndrome']))
    
    return metrics

# Evaluate on both sets
print("=== Best Parameters ===")
print(grid_search.best_params_)

train_metrics = evaluate_model(best_svm, X_train, y_train, 'Training')
test_metrics = evaluate_model(best_svm, X_test, y_test, 'Testing')

# Display results
print("\nTraining Metrics:")
print(pd.DataFrame.from_dict(train_metrics, orient='index', columns=['Value']))

print("\nTesting Metrics:")
print(pd.DataFrame.from_dict(test_metrics, orient='index', columns=['Value']))

# =============================================
# 5. ADDITIONAL ANALYSIS
# =============================================

# Cross-validation results
cv_results = pd.DataFrame(grid_search.cv_results_)
print("\nTop 5 Parameter Combinations:")
print(cv_results.sort_values('mean_test_AUC', ascending=False)[
    ['params', 'mean_test_AUC', 'mean_test_Accuracy', 'mean_test_Sensitivity']].head(5))

# Feature importance (for linear kernel)
if grid_search.best_params_['classifier__kernel'] == 'linear':
    coefs = best_svm.named_steps['classifier'].coef_[0]
    feature_names = (best_svm.named_steps['preprocessor']
                    .named_transformers_['cat']
                    .named_steps['onehot']
                    .get_feature_names_out(categorical_cols))
    feature_names = np.concatenate([non_qt_numeric, qt_numeric, feature_names])
    
    plt.figure(figsize=(10,6))
    pd.Series(coefs, index=feature_names).sort_values().plot(kind='barh')
    plt.title('SVM Feature Coefficients (Linear Kernel)')
    plt.show()
```
