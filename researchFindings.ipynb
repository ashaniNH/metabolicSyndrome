```python
#1. Import dataset
import pandas as pd
df= pd.read_csv("Metabolic Syndrome.csv")
df
```

```python
#6. Identify object type columns
object_cols = df.select_dtypes(include=['object']).columns

# Convert object type columns to category
df[object_cols] = df[object_cols].astype('category')

# Verify the conversion
print("Data types after conversion:")
print(df.dtypes)

# Optional: Show memory usage before and after conversion
print("\nMemory usage before conversion:", df.memory_usage(deep=True).sum()/1024, "KB")
```

```python
#7. Convert data types
df[['Albuminuria', 'MetabolicSyndrome']] = df[['Albuminuria', 'MetabolicSyndrome']].astype('category')
# Verify the conversion
print("Data types after conversion:")
print(df.dtypes)
```

```python
# 10. Remove missing values if they exist
if df.isnull().sum().sum() > 0:
    df_clean = df.dropna()
    print(f"\nRemoved {len(df) - len(df_clean)} rows with missing values")
    print("New dimensions after removal(train):", df_clean.shape)
else:
    print("\nNo missing values found in the dataset")
    df_clean = df.copy()
```

```python
import pandas as pd

# Define waist circumference thresholds
MALE_THRESHOLD = 101.6  # inches
FEMALE_THRESHOLD = 88.9  # inches

# Create masks for high waist circumference by gender
high_wc_males = (df_clean['Sex'] == 'Male') & (df_clean['WaistCirc'] > MALE_THRESHOLD)
high_wc_females = (df_clean['Sex'] == 'Female') & (df_clean['WaistCirc'] > FEMALE_THRESHOLD)

# Combine both genders
high_wc = high_wc_males | high_wc_females

# Count MetabolicSyndrome cases among those with high waist circumference
results = df_clean[high_wc]['MetabolicSyndrome'].value_counts().reset_index()
results.columns = ['MetabolicSyndrome', 'Count']

# Calculate percentages
total = results['Count'].sum()
results['Percentage'] = (results['Count'] / total * 100).round(1)

print("Count of Metabolic Syndrome cases in individuals with high waist circumference:")
print(results)

# Additional breakdown by gender
print("\nBreakdown by gender:")
gender_results = df_clean[high_wc].groupby(['Sex', 'MetabolicSyndrome']).size().unstack()
print(gender_results)
```

```python
import pandas as pd

# Define waist circumference thresholds
MALE_THRESHOLD = 101.6  # inches
FEMALE_THRESHOLD = 88.9  # inches

# Create masks for high waist circumference by gender
high_wc_males = (df_clean['Sex'] == 'Male') & (df_clean['WaistCirc'] < MALE_THRESHOLD)
high_wc_females = (df_clean['Sex'] == 'Female') & (df_clean['WaistCirc'] < FEMALE_THRESHOLD)

# Combine both genders
high_wc = high_wc_males | high_wc_females

# Count MetabolicSyndrome cases among those with high waist circumference
results = df_clean[high_wc]['MetabolicSyndrome'].value_counts().reset_index()
results.columns = ['MetabolicSyndrome', 'Count']

# Calculate percentages
total = results['Count'].sum()
results['Percentage'] = (results['Count'] / total * 100).round(1)

print("Count of Metabolic Syndrome cases in individuals with high waist circumference:")
print(results)

# Additional breakdown by gender
print("\nBreakdown by gender:")
gender_results = df_clean[high_wc].groupby(['Sex', 'MetabolicSyndrome']).size().unstack()
print(gender_results)
```

```python
# Filter based on Triglyceride levels ≥ 150
trigly_condition = df_clean['Triglycerides'] >= 150

# Apply the condition
filtered_df = df_clean[trigly_condition]

# Count Metabolic Syndrome and Non-Metabolic Syndrome in the filtered data
syndrome_counts = filtered_df['MetabolicSyndrome'].value_counts()

# Display result
print("Counts for individuals with Triglycerides ≥ 150 mg/dL:")
print(syndrome_counts)
```

```python
# Define the HDL condition
hdl_condition = (
    ((df_clean['Sex'] == 'Male') & (df_clean['HDL'] < 40)) |
    ((df_clean['Sex'] == 'Female') & (df_clean['HDL'] < 50))
)

# Filter the dataset based on the HDL condition
filtered_df = df_clean[hdl_condition]

# Create a crosstab of Metabolic Syndrome by Sex
output = pd.crosstab(filtered_df['Sex'], filtered_df['MetabolicSyndrome'])

# Rename columns for clarity (optional)
output.columns = ['Non-Metabolic', 'Metabolic']
output.index.name = 'Sex'

# Print breakdown
print("Breakdown by gender:")
print(output.T)
```

```python
# Filter for Fasting Blood Sugar >= 100 mg/dL
high_fbs = df_clean[df_clean['BloodGlucose'] >= 100]

# Count Metabolic Syndrome and Non-Metabolic Syndrome
fbs_counts = high_fbs['MetabolicSyndrome'].value_counts()

# Display result
print("Counts for individuals with fasting blood sugar ≥ 100 mg/dL:")
print(fbs_counts)
```

```python
# Prediabetes: 100–125 mg/dL
prediabetes = df_clean[(df_clean['BloodGlucose'] >= 100) & (df_clean['BloodGlucose'] <= 125)]
prediabetes_counts = prediabetes['MetabolicSyndrome'].value_counts()

# Type 2 Diabetes: >125 mg/dL
diabetes = df_clean[df_clean['BloodGlucose'] > 125]
diabetes_counts = diabetes['MetabolicSyndrome'].value_counts()

# Print results
print("Prediabetes (100–125 mg/dL):")
print(prediabetes_counts)

print("\nType 2 Diabetes (>125 mg/dL):")
print(diabetes_counts)
```
